### Задание 1
Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

Газоанализаторы в метро должны максимизировать Precision,  
так как необходимо уменьшить долю ложных срабатываний чтобы не создавать очереди
Металлоискатели в зеленой зоне аэропорта должны максимизировать Recall чтобы обеспечить  
максимум обнаружений

Предположим, что банк создал интеллектуальную систему выявления и блокировки подозрительных  
транзакций по картам клиентов. Тогда для транзакций ниже некой критической суммы  
(рассчитывается индивидуально, но предположим это 100 000 руб) максимизируем Precision -  
чтобы не блокировать много транзакций и не портить жизнь клиенту.  
Соотвественно, при сумме транзакций выше критической максимизируем Recall - выявляем больше подозрительных, чтобы сохранить  
деньги клиента

### Задание 2
Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

Можно попасть в ситуацию когда Recall максимален, а Precision стремится к 0. Но среднее будет равно 1/2 от Recall, что  
в численном выражении может показаться хорошим результатом, но по существу этот алгоритм просто все объекты относит к  
одному классу.
В случае F-меры такой результат невозможен - она будет стремиться к 0, при стремлении к 0 одного из показателей.
