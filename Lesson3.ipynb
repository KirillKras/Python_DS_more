### Задание 1
Для чего и в каких случаях полезны различные варианты усреднения 
для метрик качества классификации: micro, macro, weighted?

Макросреднее - метрики для каждого класса вычисляются отдельно, складываются и делятся 
на количество классов. 
Микросреднее - агрегирует вклады всех классов.  
Еслм в задаче важно просто максимизировать правильность предсказаний, то предпочтительно
микросреднее. Если классы несбалансированы, но при этом маленький класс очень важен - выявление 
малого числа больных в множестве здоровых (короновирус), то предпочтительно макросреднее так как он
не чувствителен к дисбалансу.  
При макро взвешенной средней метрики для каждого класса значение метрики умножаем на долю примеров этого класса, таким образом
учитываем дисбаланс.
Микросреднее и взвешенное макросреднее - не понимаю в чем разница,  

### Задание 2
В чём разница между моделями xgboost, lightgbm и catboost или какие их основные особенности?

xgboost не обрабатывает категориальные переменные, очень медленно работает  
lightgdm обрабатывает категориальные переменные, но тогда показывает невысокие показатели точности, быстро работает    
catboost работает быстро и с максимальными показателями эффективности именно при наличи категориальных признаков, наиболее
быстрый
